1. Pandas series

Napisz program, który utworzy obiekt Pandas Series na podstawie każdego z poniższych elementów:

mylist = list('abcedfghijklmnopqrstuvwxyz')

```python

mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

seriaWszystkich = pd.Series(data = [mylist,myarr,mydict])
seriaListy = pd.Series(data = mylist)
seriaTablicy = pd.Series(data = myarr)
seriaSlownika = pd.Series(data = mydict)

print(f'Wszystkie {type(seriaWszystkich)}: \n{seriaWszystkich}')
print(f'Lista {type(seriaListy)}: \n{seriaListy}')
print(f'Tablica {type(seriaTablicy)}: \n{seriaTablicy}')
print(f'Slownik {type(seriaSlownika)}: \n{seriaSlownika}')
```

Napisz program w którym na podstawie poniższej tabeli województw utworzysz obiekt Pandas
DataFrame o indeksie zawierającym nazwy województw i następnie:
    
● Wyznacz miasta wojewódzkie w województwach o powierzchni mniejszej niż 20000 km²

● Utwórz nowy obiekt DataFrame zawierający województwa w których liczba ludności przekracza
2000000

● Dodaj wiersz:
wielkopolskie 30 Poznań 29 826,50 3 475 323

● Usuń wiersz zawierający woj. Lubuskie

● Zastosuj funkcję pandas.DataFrame.describe() i wyznacz parametry statystyczne dla analizowanego
zbioru danych



```python
provinceSeries = pd.Series(data = ['dolnośląskie', 'kujawsko-pomorskie', 'lubelskie', 'lubuskie', 'łódzkie', 'małopolskie', 'mazowieckie']).rename("województwo")
provinceRegistrySeries = pd.Series(data = [2, 4, 6, 8, 10, 12, 14]).rename("nr rejestru")
provinceCitySeries = pd.Series(data = ['Wrocław', 'Bydgoszcz/Toruń', 'Lublin', 'Gorzów Wielkopolski / Zielona Góra', 'Łódź', 'Kraków', 'Warszawa']).rename("miasto wojewódzkie")
provinceAreaSeries = pd.Series(data = [19946.70, 17971.34, 25122.46, 13987.93, 18218.95, 15182.79, 35558.47] ).rename("powierzchnia [km²]")
provincePopulationSeries = pd.Series(data = [2904198, 2086210, 2139726, 1018084, 2493603, 3372618, 5349114]).rename("ludność")

provinceDf = pd.DataFrame({"województwo": provinceSeries, "nr rejestru": provinceRegistrySeries, "miasto wojewódzkie": provinceCitySeries, "powierzchnia [km²]": provinceAreaSeries, "ludność": provincePopulationSeries})

display(provinceDf.style.hide_index())
```

● Wyznacz miasta wojewódzkie w województwach o powierzchni mniejszej niż 20000 km²

```python
bigProvince = provinceDf[provinceDf["powierzchnia [km²]"] > 20000]
display(bigProvince.style.hide_index())
```

● Utwórz nowy obiekt DataFrame zawierający województwa w których liczba ludności przekracza 2000000

```python
populatedProvince = provinceDf[provinceDf["ludność"] > 2000000]
display(populatedProvince.style.hide_index())
```

```python
wlkpProvince = pd.Series({"województwo": "wielkopolskie", "nr rejestru": 30, "miasto wojewódzkie": "Poznań", "powierzchnia [km²]": 29826.50, "ludność": 3475323})
provinceDf = provinceDf.append(wlkpProvince, ignore_index=True)
display(provinceDf.style.hide_index())
```

● Posortuj dane po liczbie ludności malejąco

```python
provinceDf = provinceDf.sort_values("ludność", ascending=False)
display(provinceDf.style.hide_index())
```

● Przesuń kolumnę “miasto wojewódzkie” na ostatnie miejsce

```python
#provinceDf = provinceDf[['województwo', 'nr rejestru', 'powierzchnia [km²]', 'ludność', 'miasto wojewódzkie']]
display(provinceDf.style.hide_index())
citiesDf = provinceDf.pop('miasto wojewódzkie')
display(provinceDf.style.hide_index())
provinceDf['miasto wojewódzkie'] = citiesDf
display(provinceDf.style.hide_index())
```
