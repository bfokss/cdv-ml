<b>7.</b> Model training

```python
# Preparing dataset

X = data.drop('blueWins', axis=1).to_numpy()
display(X)
```

```python
# Preparing labels

y = data.loc[:, 'blueWins'].to_numpy()
display(y)
```

```python
# Splitting dataset into training dataset and test dataset

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12345) 
```

```python
# Test model method

def test_model(classifier, feature_vector_train, label, feature_vector_valid):
    # Train model
    classifier.fit(feature_vector_train, label)
    
    # Generate predictions for test dataset
    predictions = classifier.predict(feature_vector_valid)
    
    # Evaluate model
    scores = list(metrics.precision_recall_fscore_support(predictions, y_test))
    score_vals = [
        scores[0][0],
        scores[1][0],
        scores[2][0]
    ]
    score_vals.append(metrics.accuracy_score(predictions, y_test))
    return score_vals
```

```python
# MODEL 1 - Logistic Regression
accuracy = test_model(linear_model.LogisticRegression(), X_train, y_train, X_test)
accuracy_compare = {'LR': accuracy}
print ("LR: ", accuracy)
```

```python
# MODEL 2 - Support Vector Machine
accuracy = test_model(svm.SVC(), X_train, y_train, X_test)
accuracy_compare['SVM'] = accuracy
print ("SVM" , accuracy)
```

```python
# MODEL 3 - Random Forest Tree 
accuracy = test_model(ensemble.RandomForestClassifier(), X_train, y_train, X_test)
accuracy_compare['RF'] = accuracy
print ("RF: ", accuracy)
```

```python
#Models comparison

df_compare = pd.DataFrame(accuracy_compare, index = ['precision', 'recall', 'f1 score', 'accuracy'])
df_compare.plot(kind='bar')
```

```python
plt.close('all')
```

<b>As above comparison shows. Best model would be Logistic regression</b>

```python
# Train model method

def train_model(classifier, feature_vector_train, label, feature_vector_valid):
    # Train model
    classifier.fit(feature_vector_train, label)
    
    # Generate predictions for test dataset
    predictions = classifier.predict(feature_vector_valid)
    
    # Save model to .pickle file
    with open('lol_classifier.pickle', 'wb') as handle:
        pickle.dump(classifier, handle)
    
    # Evaluate model
    scores = list(metrics.precision_recall_fscore_support(predictions, y_test))
    score_vals = [
        scores[0][0],
        scores[1][0],
        scores[2][0]
    ]
    score_vals.append(metrics.accuracy_score(predictions, y_test))
    return classifier
```

```python
# Prepare data for check

test_data = data.iloc[55]
blueWins = test_data.blueWins
test_data = test_data.drop(labels='blueWins')
display(test_data)
print(f"Has blue team won?[0-No / 1-Yes]: {blueWins}")
```

```python
# Train model and assign it to variable

model = train_model(linear_model.LogisticRegression(), X_train, y_train, X_test)
```

```python
# Load model from file

loaded_model = pickle.load(open('lol_classifier.pickle', 'rb'))
```

```python
# Predict with both models(same model) 

print("Loaded model: ")
loaded_result = loaded_model.predict([test_data])
print(loaded_result)
print("Variable model: ")
result = model.predict([test_data])
print(result)
print("Actual match result:")
print(blueWins)
```
