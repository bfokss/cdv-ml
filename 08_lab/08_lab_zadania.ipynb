```python
blueColumns = data.filter(like='blue').columns
blueData = data[blueColumns]
display(blueData)
```

```python
redColumns = data.filter(like='red').columns
redData = data[redColumns]
display(redData)
```

```python
import seaborn as sns
correlations = redData.corr()
fig, ax = plt.subplots(figsize=(20, 20))
colormap = sns.color_palette("BrBG", 10)
sns.heatmap(correlations,
     cmap=colormap,
     annot=True,
     fmt=".2f")

plt.show()
```

```python
plt.close('all')
```

```python
correlations = blueData.corr()
fig, ax = plt.subplots(figsize=(20, 20))
colormap = sns.color_palette("BrBG", 10)
sns.heatmap(correlations,
     cmap=colormap,
     annot=True,
     fmt=".2f")

plt.show()
```
